### Implementing Regression using KNN

```python
# importing required libraries

import pandas as pd
import numpy as np
```

```python
# creating a dataset

from sklearn.datasets import make_regression
X,y=make_regression(
    n_samples=1000,
    n_features=4
)
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
# splitting our dataset into training and testing splits

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=7)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# fitting our data to the model and finding our best 'K' value

# creating our KNN Regression model

from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

lowest_mse=0
k_best=1

for i in range(1,11):

    test_knn_regressor=KNeighborsRegressor(n_neighbors=i)

    test_knn_regressor.fit(X_train, y_train)

    test_preds=test_knn_regressor.predict(X_test)

    mse=mean_squared_error(y_test, test_preds)

    if(i==1):
        lowest_mse=mean_squared_error(y_test, test_preds)

    if(mse<lowest_mse):

        k_best=i
        lowest_mse=mse

print(f"K:{k_best}\nMean Squared Error:{lowest_mse}")

```

```python
knn_regressor=KNeighborsRegressor(n_neighbors=6)
knn_regressor.fit(X_train, y_train)
y_preds=knn_regressor.predict(X_test)
mse=mean_squared_error(y_test, y_preds)
print(mse)
```
