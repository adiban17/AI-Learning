## Binary Classification using Logistic Regression

```python
# importing basic libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# creating a dataset

from sklearn.datasets import make_classification
X,y=make_classification(n_features=10, n_samples=1000, n_classes=2,  random_state=42)
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
# splitting the data into train, test sets

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y)
```

```python
# training our Logistic Regression Model

from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
model.predict_proba(X_test)
```

```python
# validating the performance of our model using Performance Metrics

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
accuracy=accuracy_score(y_test, y_pred)
cm=confusion_matrix(y_test, y_pred)
p_score=precision_score(y_test, y_pred)
r_score=recall_score(y_test, y_pred)
print(f"Accuracy: {accuracy}\nConfusion Matrix: \n{cm}\nPrecision Score: {p_score}\nRecall Score: {r_score}")
```

```python
# checking the classification report of our model

from sklearn.metrics import classification_report
cr=classification_report(y_test, y_pred)
print(f"Classification Report:\n{cr}")
```

```python
# Hyper-Parameter Evaluation

log_reg=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
params=dict(penalty=penalty,C=c_values, solver=solver)
```

```python
# Hyper-Parameter Tuning: Grid-Search CV

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold(n_splits=5, shuffle=False)
grid=GridSearchCV(estimator=log_reg, param_grid=params, scoring='accuracy',cv=cv, n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
# evalidating the performance of our Grid CV model using Performance Metrics

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
accuracy=accuracy_score(y_test, y_pred)
cm=confusion_matrix(y_test, y_pred)
p_score=precision_score(y_test, y_pred)
r_score=recall_score(y_test, y_pred)
print(f"Accuracy: {accuracy}\nConfusion Matrix: \n{cm}\nPrecision Score: {p_score}\nRecall Score: {r_score}")
```

```python
# checking the classification report of our Grid CV model

from sklearn.metrics import classification_report
cr=classification_report(y_test, y_pred)
print(f"Classification Report:\n{cr}")
```

```python
# Hyper-Parameter Tuning: Randomized Search CV

from sklearn.model_selection import RandomizedSearchCV
rscv=RandomizedSearchCV(estimator=model, param_distributions=params, scoring='accuracy', n_jobs=-1, cv=5)
```

```python
rscv.fit(X_train, y_train)
```

```python
rscv.best_params_
```

```python
rscv.best_score_
```

```python
y_pred=rscv.predict(X_test)
```

```python
# evalidating the performance of our Randomized Search CV model using Performance Metrics

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
accuracy=accuracy_score(y_test, y_pred)
cm=confusion_matrix(y_test, y_pred)
p_score=precision_score(y_test, y_pred)
r_score=recall_score(y_test, y_pred)
print(f"Accuracy: {accuracy}\nConfusion Matrix: \n{cm}\nPrecision Score: {p_score}\nRecall Score: {r_score}")
```

```python
# checking the classification report of our Randomized Search CV model

from sklearn.metrics import classification_report
cr=classification_report(y_test, y_pred)
print(f"Classification Report:\n{cr}")
```
