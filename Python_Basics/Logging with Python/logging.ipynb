### Python Logging
Logging is a crucial aspect of any application, providing a way to track events, errors, and operational information. Python's built-in logging module provides a flexible framework for emitting log messages from Python programs.

```python
import logging

# configure the basic logging settings
logging.basicConfig(level=logging.DEBUG)

# different logging messages with different severity levels
logging.debug("This is a debigging message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical  message")
```

#### Log Levels:
Python's logging modules have several log levels indicating the severity of events. The default levels are:<br>
<br>
* DEBUG: Detailed information, typically of interest only when diagonising problems.<br>
* INFO: Confirmation that things are working as expected.<br>
* WARNING: An indication that something unexpected happened or indicative of some problem in the near future(eg:-disc space low). Software is still running fine.<br>
* ERROR: Due to a more serious problem, the software is not able to perform some problem.<br>
* CRITICAL: A very serious error, indicating that the program itself might not continue running.


```python
import logging

# configuring logging
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)

# different logging messages with different severity levels
logging.debug("This is a debigging message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical  message")
```
