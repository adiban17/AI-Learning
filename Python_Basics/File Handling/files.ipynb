```python
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
# Overwriting a File
with open('example.txt','w') as file:
    file.write("I am overwriting this file!")
    file.write("This seems pretty cool.")
    file.write("This is not on a new line\n")
    file.write("This is a new line")
```

```python
# Reading te overwritten file to check the changes made
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
# Appending new text to a file
with open('example.txt','a+') as file:
    file.write("Right now i am using the '.write' function to write new stuff within a file\n")
    file.write("But this new stuff is not getting overwritten over the old stuff. This is being appended.\n")
    file.write("I still love food though.ðŸ˜‹")
    for line in file:
        print(line.strip())
```
