## Generators

Generators are a simpler way to create iterators. They use the yield keyword to produce a set of values lazily, which means that they create a set of values  on the fly and do not store them in memory.

```python
def square(n):
    for x in range(n):
        return(x**2)
```

```python
square(3)
```

```python
def square(n):
    for x in range(n):
        yield(x**2)
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
a=square(10)
print(a)
```

```python
try:
    print(next(a))
except Exception as ex:
    print(ex)
```

```python
# Real World Example of using Generators: Reading Large Files
def read_file(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line
```

```python
read_file('largefile.txt')
```

```python
for line in read_file(file_path='largefile.txt'):
    print(line.strip())
```
