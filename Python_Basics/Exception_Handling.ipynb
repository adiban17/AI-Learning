## Exception Handling

try--> contains risky code that might have errors<br>
except--> executes if any/specified error occurs in the try block <br>
else--> executes if no exception is detetcted<br>
finally--> always executes no matter what

```python
a=b
```

```python
try:
    a=b
except:
    print("Variable is not defined")
```

```python
try:
    a=b
except Exception as ex:
    print(ex)
```

```python
try:
    division=3/0
    print(division)
except Exception as ex:
    print(ex)
    print("Enter a divior other than 0")
```

```python
try:
    num=int(input("Enter a Number: "))
    quo=10/num
except ZeroDivisionError:
    print("Enter a Number other than 0")
except ValueError:
    print("Enter a Number")
except Exception as ex:
    print(ex)
```

```python
try:
    num=int(input("Enter a Number: "))
    quo=10/num
except ZeroDivisionError:
    print("Enter a Number other than 0")
except ValueError:
    print("Enter a Number")
except Exception as ex:
    print(ex)
else:
    print(quo)
```

```python
try:
    num=int(input("Enter a Number: "))
    quo=10/num
except ZeroDivisionError:
    print("Enter a Number other than 0")
except ValueError:
    print("Enter a Number")
except Exception as ex:
    print(ex)
else:
    print(quo)
finally:
    print("Execution Completed")
```

```python
try:
    file=open('example1.txt','r')
except FileNotFoundError as ex:
    print(ex)
except Exception as ex:
    print(ex)
finally:
    if 'file' in locals() or not file.closed():
        file.close()
        print("File Closed")
```
