## Iterators

Iterators are adavnced Python concepts that allow for efficient looping and memory management. Iterators provide a way to access collection elements without exposing the underlying structure.

```python
my_list=[1,2,3,4,5]
type(my_list)
```

```python
# creating an iter object
iterator=iter(my_list)

type(iterator)
```

```python
next(iterator)
```

```python
next(iterator)
```

```python
name='Aditya'

iterator=iter(name)
type(iterator)
```

```python
try:
    print(next(iterator))
except StopIteration:
    print('No more elements to be iterated')
```
