```python
class Dog:
    # contructor
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def bark(self):
        print(f"{self.name} says bhow bhow!")

dog1=Dog("Half Ticket",1/2)
dog1.bark()

dog2=Dog("Dogesh Deshpande",4)
dog2.bark()
```

```python
# Modelling a Bank Account

class Bank:

    def __init__(self, name, balance):
        self.name=name
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print("Your bank balance is:",self.balance)

    def withdraw(self,amount):
        if self.balance>amount:
            self.balance-=amount
            print("Money Withdrawn, new balance is:",self.balance)
        else:
            print("Insufficient Funds")

    def get_balance(self):
        print(f'Your Bank Balance is {self.balance}')

customer1=Bank("Choti Lulli", 15)
customer1.withdraw(11)
customer1.withdraw(10)
```
