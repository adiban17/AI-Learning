### Encapsulation
Wrapping data(variables) and methods(functions) as a single unit. Prevents interference and data leaks.

```python
class Person:
    def __init__(self,name,age):
        self.name=name  # Public Variable
        self.age=age    # Public Variable

person1=Person('Aditya',20)
print(person1.name)
dir(person1)
```

```python
class Person:
    def __init__(self,name,age):
        self.__name=name  # Private Variable (can only be called within the class)
        self.__age=age    # Private Variable (can only be called within the class)

person2=Person('Han Khan',20)
print(dir(person2))
try:
    person2.age()
except Exception as ex:
    print(ex)
```

```python
class Person:
    def __init__(self,name,age):
        self._name=name  # Protected Variable (can only be called within the class and inherited classes)
        self._age=age    # Protected Variable (can only be called within the class and inherited classes)

class Student(Person):
    def __init__(self, name, age):
        super().__init__(name, age)

person3=Student('Half Ticket',14)
print(person3._name)
```

```python
# Encapsulation using Getter and Setter Methods

class Person:
    def __init__(self,name,age):
        self.__name=name  # Private Variable (can only be called within the class)
        self.__age=age    # Private Variable (can only be called within the class)

    # Getter Method to get name
    def get_name(self):
        return self.__name
    
    # Setter Method to set name
    def set_name(self,name):
        self.__name=name

person4=Person('Adi',19)
print(person4.get_name())
person4.set_name('Han')
print(person4.get_name())

        
# Getter and Setter Methods are used to view/change the values of private variables from outside the class
```
