### Abstraction
Abstraction is the concept of hiding complex implemetation details and showing only the necessary features of the object. This helps in reducing the complexity of the code.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    def drive(self):
        print("This vehicle is driving")

    @abstractmethod
    def engine_start(self):
        pass

class Car(Vehicle):

    def engine_start(self):
        return("Car Engine Started")

vehicle1=Car()
vehicle1.engine_start()
```
