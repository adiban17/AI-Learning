```python
class Car():
    
    def __init__(self,brand,engine,doors):
        self.brand=brand
        self.engine=engine
        self.doors=doors

    def drive(self):
        print(f"The driver is driving a {self.engine} {self.brand}")

car1=Car('BMW','Inline 6', 4)
car1.drive()
```

```python
# Single Inheritance
class Tesla(Car):

    def __init__(self, brand, engine, doors,is_selfdriving):
        super().__init__(brand, engine, doors)
        self.is_selfdriving=is_selfdriving

    def selfdriving(self):
        print(f'Tesla supports self driving:{self.is_selfdriving}')

car2=Tesla('Tesla','electric',2, True )
car2.drive()
car2.selfdriving()
```

```python
# Multiple Inheritance
class Pet:
    def __init__(self,name,type):
          self.name=name
          self.type=type
        
class Sound(Pet):
    def __init__(self, name, type, sound):
        super().__init__(name, type)
        self.sound=sound

    def ssound(self):
         print(f"{self.name} says {self.sound}")

class Owner( Sound):
     
    def __init__(self, name, type,sound,owner):
        super().__init__(name, type, sound)
        
        self.owner=owner

    def Print_owner_name(self):
        print(f"Owner:{self.owner}")

pet1=Owner('Tommy','Dog','Woof','S.M.H.K.')
pet1.ssound()


        

     

    
```
