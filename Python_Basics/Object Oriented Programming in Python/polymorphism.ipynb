### Method Overrriding:
Same method name but different functionality

```python
class Animal:
    def speak(self):
        print("Sound of the Animal")

class Dog(Animal):
    def speak(self):
        print("Woof")

class Cat(Animal):
    def speak(self):
        print("Meow")

# Function that demonstrates polymorphism
def animal_speak(animal_name):
    animal_name.speak()

cat=Cat()
dog=Dog()
cat.speak()
dog.speak()

animal_speak(dog)
```

```python
class Shape:
    def Area():
        print("Area of the Shape")

class Rectangle(Shape):

    def __init__(self,width,height):
        self.width=width
        self.height=height

    def Area(self):
        return(self.width*self.height)
    
class Circle(Shape):

    def __init__(self,radius):
        self.radius=radius

    def Area(self):
        return(3.14*(self.radius**2))
    
shape1=Circle(234)
shape2=Rectangle(20,12)
print(shape1.Area())
print(shape2.Area())
```

```python
# Polymorphism with Abstract Base Class

# Similar to Interfaces in other programming languages. They are used to define classes for  objects with common characteristics

from abc import ABC,abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

# Derived Class 1
class Car(Vehicle):
    def start_engine(self):
        return("Car engine started")
    
# Derived Class 2
class Motorcycle(Vehicle):
    def start_engine(self):
        return("Motorcycle Engine Started")
    
# Create Objects of Car and Motorcycle
vehicle1=Car()
vehicle2=Motorcycle()
print(vehicle1.start_engine())
print(vehicle2.start_engine())

```
