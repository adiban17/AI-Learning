**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
year=2012
if year%4==0 and year%100!=0:
    print('Leap Year')
elif year%400==0:
    print('Leap Year')
else:
    print('Not a Leap Year')
```
