#### Handling Imbalanced Datasets:
1. Upsampling
2. Downsampling

```python
import pandas as pd
import numpy as np
```

```python
# creating an imbalanced dataset
dataset={'Age':[12,15,23,43,55,21,22,34,90,54],'Disease':['Yes','Yes','No','Yes','Yes','Yes','No','Yes','Yes','No']}
data=pd.DataFrame(dataset)
data
```

```python
data['Disease'].value_counts()
```

```python
# upsampling the data
data_majority=data[data['Disease']=='Yes']
data_minority=data[data['Disease']=='No']

from sklearn.utils import resample
data_minority_resampled=resample(data_minority, replace=True , # sample with replacement
    n_samples=len(data_majority),
    random_state=42 )

data_minority_resampled
```

```python
data_minority_resampled.shape
```

```python
balanced_dataset=pd.concat([data_minority_resampled, data_majority])
balanced_dataset
```

```python
balanced_dataset['Disease'].value_counts()
```

```python
# downsampling
data_majority=data[data['Disease']=='Yes']
data_minority=data[data['Disease']=='No']

from sklearn.utils import resample
data_majority_resampled=resample(data_majority, replace=False ,
    n_samples=len(data_minority),
    random_state=42 )

data_majority_resampled
```

```python
balanced_dataset=pd.concat([data_majority_resampled, data_minority])
balanced_dataset
```

```python
balanced_dataset['Disease'].value_counts()
```
