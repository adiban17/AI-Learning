```python
import numpy as np
```

```python
np.random.seed(69)
head,tail=0,0
for x in range(0,1000000):
    res=np.random.randint(0,2)
    if res==0:
        head+=1
    else:
        tail+=1

print("Head:",((head/1000000)*100))
print("Tail:",((tail/1000000)*100))
```
