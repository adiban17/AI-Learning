### Zomato Dataset EDA

```python
# importing basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the dataset
df = pd.read_csv('zomato.csv', encoding='latin-1')
```

```python
df.isnull().sum()
```

```python
# importing country code dataset
country_code=pd.read_excel('Country-Code.xlsx')
country_code.head()
```

```python
# Merging the Country Codes with the Main Dataset
new_df=pd.merge(df,country_code,on='Country Code',how='left')
new_df.head()
```

```python
new_df.shape
```

```python
tranactions_per_country=new_df.value_counts('Country')
country_names=tranactions_per_country.index
tranactions_per_country
```

```python
country_names
```

```python
# creating a pie-chart for the transactions per country for top 3 countries
plt.pie(tranactions_per_country[:3],labels=country_names[:3],autopct='%1.2f%%');
```

##### Observation 1:
The top 3 countries with the highest number of transactions are India, USA and UK.

```python
ratings=pd.DataFrame(new_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index())
ratings
```

```python
ratings.rename(columns={0:'Rating Count'},inplace=True)
ratings.head()
```

#### Observation 2:
- 0.0--->White--->Not Rated
- 0.1-2.4---->Red--->Poor
- 2.5-3.4---->Orange--->Average
- 3.5-3.9---->Yellow--->Good
- 4.0-4.4---->Green--->Very Good
- 4.5-4.9---->Dark Green--->Excellent

```python
# Plotting a Bar Chart for the ratings
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','darkgreen'])
```
