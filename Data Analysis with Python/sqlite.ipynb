### SQL and SQLite
SQL (sructured query language) is the standard language for managing and manipulating relational databases. SQLite is a self contained serverless and zero configuration database engine that is widely used for embedded database systems.  

```python
import sqlite3
```

```python
# connect to an sql database
connection=sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
# create a table
cursor.execute('''
Create table if not exists employees(
    id Integer primary key,
    name text not null,
    age integer,
   department text)
''')

# commit the changes
connection.commit()
```

```python
cursor.execute('''
    Select * from Employees
''')
```

```python
# inserting data in sqlite3 database
cursor.execute('''
Insert Into Employees(name,age,department)
               Values('Bob',25,'Engineering')
''')

cursor.execute('''
Insert Into Employees(name,age,department)
               Values('David',32,'Finance')
''')

cursor.execute('''
Insert Into Employees(name,age,department)
               Values('Charles',17,'Plumber')
''')

connection.commit()
```

```python
# query the data from the table
cursor.execute('Select * from employees')
rows=cursor.fetchall()

# print the queries data
for row in rows:
    print(row)
```

```python
# updating the data
cursor.execute('''
Update employees
               set age=19
               where name="Charles"
''')
connection.commit()
```

```python
# query the data
cursor.execute(' select * from employees')
rows=cursor.fetchall()

# print the queries data
for row in rows:
    print(row)
```

```python
# deleting data
cursor.execute('''
delete from employees
               where name="David"
''')

connection.commit()
```
