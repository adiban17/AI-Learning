```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

# creating a line plot
plt.plot(x,y)
plt.xlabel('number')
plt.ylabel('square')
plt.title('relation of numbers with their squares')
plt.show()
```

```python
# customizing our plot
plt.plot(x,y,color='red',linestyle='--',linewidth=2, marker='o', markersize=10)
plt.grid(True)
plt.show()
```

```python
# creating subplots
plt.figure(figsize=(9,5))

plt.subplot(1,1,1)
plt.plot(x,y,color='green',marker='o')
plt.title('Plot 1')

plt.subplot(1,2,2)
plt.plot(x,y,color='red',marker='o')
plt.title('Plot 2')

plt.subplot(2,1,1)
plt.plot(x,y,color='red',marker='o')
plt.title('Plot 3')
```

```python
# Plotting a bar graph
player_name=['Messi','Ronaldo','Benzema','Neymar','Mbappe']
rating=[97,99,94,95,93]

plt.bar(player_name,rating,color='blue')
plt.xlabel('Player Name')
plt.ylabel('Rating')
plt.title('FIFA Ratings')
plt.show()
```

```python
# plotting a histogram
data=[1,2,2,2,4,4,4,3,3,3,3,3,3,3,3,5,5,5,5,5]

plt.hist(data,bins=5,edgecolor='black',color='orange')
plt.show()
```

```python
# plotting a scatter plot
x=[1,2,3,4,5]
y=[6,7,8,9,10]

plt.scatter(x,y,color='blue',marker='o')
plt.show()
```

```python
# creating a pie chart
labels=['a','b','c','d','e']
sizes=[20,38,20,37,68]
colors=['green','orange','coral-blue','black','pink']
explode=(0.2,0,0,0,0)  # makes the 1st pie pop out

plt.pie(sizes,labels=labels,autopct="%1.1f%%",explode=explode, shadow=True)
```

```python
# plotting sales data
import pandas as pd
```

```python
df=pd.read_csv('sales_data_sample.csv',encoding='unicode_escape')
df.head()
```

```python
df.describe()
```

```python
total_sales_country=df.groupby('COUNTRY')['QUANTITYORDERED'].sum()
total_sales_country
```

```python
total_sales_country.plot(kind='bar',color='green')
plt.ylabel('Quantity Ordered')
plt.title('Sales Data')
plt.show()
```

```python
total_sales_msrp=df.groupby('COUNTRY')['MSRP'].sum()
total_sales_msrp
```

```python
total_sales_msrp.plot(kind='bar',color='red')
plt.ylabel('MSRP')
plt.title('Sales Data by MSRP')
plt.show()
```
