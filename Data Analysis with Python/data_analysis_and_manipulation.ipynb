### Data Manipulation and Analysis with Pandas
Data Manipulation and analysis are key tasks in any data science or data analysis project. Pandas provide a wide range of function for data analysis and manipulation. making it easy to clean, transform and extract insights from data.

```python
import pandas as pd
```

```python
data=pd.read_csv('Orders.csv')
data.head()
```

```python
data.tail()
```

```python
data.dtypes     # object=string
```

```python
data.describe()
```

```python
# Checkimg if there any any missing values in the dataset
data.isnull().sum()
```

```python
# filling missing names with 0
data['CustomerName(filled NA)']=data['CustomerName'].fillna('-')
data
```

```python
data.drop('CustomerName',axis=1,inplace=True)
```

```python
data.head()
```

```python
# Renaming Columns
data=data.rename(columns={'CustomerName(filled NA)':'Customer Name'})
data.head()
```
