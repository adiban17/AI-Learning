## Pandas
Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrames. Series is a 1-D array like object, while DataFrame is a 2-D heterogeneous tabular data structure (like excel rows and columns).

#### Importing Pandas

```python
# importing pandas
import pandas as pd
```

#### Series

```python
# Series 
data=[1,2,3,4,5]

# Series from list
series_from_list=pd.Series(data)
series_from_list
```

```python
# Series from Dictionary
dictionary_data={'a':1,'b':2,'c':3}
series_from_dictionary=pd.Series(dictionary_data)
series_from_dictionary
```

```python
# Dictionary Seris from merging 2 lists
data=['adi','han','bheltu']
index=[1,2,3]
new_series=pd.Series(data,index=index)
new_series
```

#### DataFrames

```python
# DataFrames

# Creating a DataFrame from dicitonary of list
data={
    'name':['adi','han','bheltu','chomesh','htrt3'],
    'age':[20,19,23,21,27],
    'city':['Mumbai','Kolkata','Madrid','Zagreb','Naihati']
}
df=pd.DataFrame(data)
df
```

```python
# Creating a DataFrame from list of Dictionaries
data=[
    {'name':'adi','age':20,'city':'mumbai'},
    {'name':'han','age':19,'city':'kolkata'},
    {'name':'bheltu','age':23,'city':'madrid'},
    {'name':'chomesh','age':21,'city':'zagreb'},
    {'name':'htrt3','age':27,'city':'naihati'},
]
df=pd.DataFrame(data)
df
```

```python
# Acessing elements in a DatFrame
df.loc[0]
```

```python
df.iloc[0]
```

```python
df.iloc[1][2]
```

```python
# Accessing a specific element using at
df.at[2,'city']
```

```python
df.at[0,'name']
```

```python
# Accessing a specific element using iat
df.iat[0,0]
```

```python
df.iat[2,1]
```

```python
# Adding a column
print(df)

df['Salary']=[1000,2340,970,1300,20383]
print(df)
```

```python
# Removing a Column
df.drop('Salary',axis=1,inplace=True)
df
```

```python
# Add 1 to each age in the column
df['age']=df['age']+1
df
```
