```python
import pandas as pd
```

```python
# importing our dataset

messages=pd.read_csv('SMSSpamCollection.csv')
messages.head()
```

### Data Cleaning and Preprocessing

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
porterstemmer=PorterStemmer()
```

```python
corpus=[]

for i in range(len(messages)):
    review=re.sub('[^a-zA-Z]', ' ', messages['Message'][i])
    review=review.lower()
    review=review.split()
    review=[porterstemmer.stem(word) for word in review if word not in stopwords.words()]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# creating our Bag of Words Model

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500)
cv_binary=CountVectorizer(max_features=2500, binary=True)   # creates a binary BOW
```

```python
X=cv.fit_transform(corpus).toarray()
```
