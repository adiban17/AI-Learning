```python
paragraph="Time is a deeply abstract and universal concept that governs the very fabric of our reality, yet it remains one of the most elusive and mysterious phenomena we encounter. It flows consistently forward—seemingly immutable and constant—yet our perception of it is anything but linear or uniform. Philosophers have debated whether time is an inherent dimension of the universe or simply a mental construct, a way for consciousness to interpret change. In physics, time is treated as both a measurable entity, appearing as the fourth dimension in Einstein’s theory of relativity, and as a relative experience, where it bends and stretches in the presence of gravity and speed. Psychologically, time is fluid—moments of joy seem to vanish in the blink of an eye, while dread can make minutes stretch into eternity. Time also plays a vital role in shaping meaning: memories root us in the past, aspirations pull us toward the future, and mindfulness grounds us in the fleeting present. Whether seen as a river we float through or a canvas we paint with our actions, time remains both a constraint and a canvas for all of existence—the inexorable rhythm by which all life dances forward into the unknown."
```

```python
import nltk
from nltk.corpus import stopwords
sentences=nltk.sent_tokenize(paragraph)
sentences
```

```python
for i in range(len(sentences)):
    sentences[i]=nltk.word_tokenize(sentences[i])
    sentences[i]=[word for word in sentences[i] if word.lower() not in set(stopwords.words())]
sentences
```

```python
for i in range(len(sentences)):
    print(nltk.pos_tag(sentences[i]))
```
