```python
document="Aditya is a boy who plays with a toy."
```

```python
import nltk
words=nltk.word_tokenize(document)
words
```

```python
tag_elements=nltk.pos_tag(words)
tag_elements
```

```python
nltk.ne_chunk(tag_elements).draw()
```
