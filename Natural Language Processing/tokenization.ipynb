```python
corpus="Hello, I am Aditya. I am 20 years old. I am a boy. I love Football."
```

```python
corpus
```

```python
# Tokenizing our courpus
# Corpus ---> Sentences

from nltk.tokenize import sent_tokenize
documents = sent_tokenize(corpus)
```

```python
documents
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
# Sentences ---> Words

from nltk.tokenize import word_tokenize, wordpunct_tokenize
words=word_tokenize(corpus)
```

```python
words
```

```python
for sentence in documents:
    words=word_tokenize(text=sentence)
    print(words)
```

```python
for sentence in documents:
    words=wordpunct_tokenize(sentence)
    print(words)
```

```python
from nltk import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()

print(tokenizer.tokenize(corpus))
```

word_tokenizer: tokenizes every single word

wordpunct_tokeizer: tokenizes every single word along with punctuatio marks other than sentrence ending punctuation marks

TreeBankTokenizer: does not take full stop as an individual word except the last full stop
