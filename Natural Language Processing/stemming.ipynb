```python
words=['go', 'goa', 'gone', 'going', 'programs', 'programming', 'programmer']
```

#### Porter Stemmer

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
for word in words:
    print(stemmer.stem(word))
```

```python
stemmer.stem('Congratulate')
```

### RegexpStemmer Class

```python
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('doing')
```

```python
reg_stemmer.stem('ingdoing')
```

```python
reg_stemmer.stem('doinginging')
```

```python
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing|s$|e$|able$', min=4)
reg_stemmer.stem('doinginging')
```

```python
reg_stemmer.stem('ingdoing')
```

### SnowBall Stemmer

```python
from nltk.stem import SnowballStemmer
stemmer=SnowballStemmer(language='english')
for word in words:
    print(stemmer.stem(word))
```
