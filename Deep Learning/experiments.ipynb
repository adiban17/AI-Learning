```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pickle
```

```python
# loading the data-set

data=pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data.info()
```

```python
# dropping irrelevant features

data.drop(columns=['RowNumber','CustomerId','Surname'], axis=1, inplace=True)
data.head()
```

```python
# encoding categorical variables

from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
data['Gender']=le1.fit_transform(data['Gender'])
data
```

```python
data['Geography'].value_counts()
```

```python
data.info()
```

```python
# OHE the Geography column

from sklearn.preprocessing import OneHotEncoder
ohe_encoder=OneHotEncoder()
geo_encoded=ohe_encoder.fit_transform(data[['Geography']]).toarray()
geo_encoded
```

```python
print(ohe_encoder.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df=pd.DataFrame(data=geo_encoded, columns=ohe_encoder.get_feature_names_out(['Geography']))
geo_encoded_df.head()
```

```python
# combining OHE encoded columns with our DataFrame

data=pd.concat([data, geo_encoded_df], axis=1)
```

```python
data.drop(columns=['Geography'], axis=1,inplace= True)
```

```python
data.head()
```

```python
# saving the encoder and scaler

with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(le1, file)

with open('onehotencoder_Geography', 'wb') as file:
    pickle.dump(ohe_encoder, file)
```

```python
# dividing our data into independent and dependent features

X=data.drop(columns=['Exited'], axis=1)
y=data['Exited']
```

```python
data['Exited'].value_counts()
```

```python
# Splitting our data into traiing and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=23)

# Standardising our Data
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler, file)
```
